# F1 driver dataset preprocessing 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.preprocessing import LabelEncoder
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('F1Drivers_Dataset.csv', encoding_errors='ignore')
```

```python
df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
# Pole Rate, convert it into percentage.

df['Pole_Rate'] = df['Pole_Rate'].apply(lambda x: x*100)
```

```python
df.head(20)
```

```python
df['Pole_Rate'] = round(df['Pole_Rate'], 2)
```

```python
df['Pole_Rate'].head(20)
```

```python
df['Start_Rate'] = df['Start_Rate'].apply(lambda x: x*100)
df['Start_Rate'] = round(df['Start_Rate'], 2)
df.head()
```

```python
df['Win_Rate'] = df['Win_Rate'].apply(lambda x: x*100)
df['Win_Rate'] = round(df['Win_Rate'], 2)
df.head()
```

```python
df['Podium_Rate'] = df['Podium_Rate'].apply(lambda x: x*100)
df['Podium_Rate'] = round(df['Podium_Rate'], 2)
df.head()
```

```python
df['FastLap_Rate'] = df['FastLap_Rate'].apply(lambda x: x*100)
df['FastLap_Rate'] = round(df['FastLap_Rate'], 2)
df.head()
```

```python
df.head(20)
```

```python
df.to_csv("f1_cleaned.csv", index=False)
```
